**Missing Value Kolom Income**\
Pertama, kita duplikat dulu dataframe-nya

kedua, cari nilai rata-rata dari kolom Income

Ketiga, Isikan nilai kosong di kolom Income dengan nilai rata-rata yang didapatkan

Keempat, cek kembali apakah masih terdapat missing value di kolom Income

```python
#Langkah 1 duplikat dataframe
df1 = df
df1
```

```python
# Langkah 2
rata_income = df1["Income"].mean()
rata_income
```

```python
pembulatan = (round(rata_income, 1))
pembulatan
```

```python
# Langkah 3
df1['Income'] = df1['Income'].fillna(pembulatan)
```

```python
# Langkah 4
df1['Income'].isna().sum()
```
